# Defining a Mean-Reversion Strategy (Bollinger Bands) (Part 1)

__Mean Reversion__: Financial Instruments are from time to time overbought / oversold and revert back to mean prices. 

__Bollinger Bands__: Consists of a SMA (e.g. 30) and Upper and Lower Bands +- (2) Std Dev away from SMA.

```python
SMA = 30 # Simple moving avarage
dev = 2 # deviations
```

```python
# Compute Simple Moving avarage to the price
data["SMA"] = data["price"].rolling(SMA).mean()
```

```python
# Plot the price, and its SMA
data[["price", "SMA"]].plot(figsize = (12, 8))
plt.show()
```

```python
# ZOOM into 2019-08 
data.loc["2019-08", ["price", "SMA"]].plot(figsize = (12, 8))
plt.show()
```

```python
# Rolling std (hint)
data["price"].rolling(SMA).std()
```

```python
# Plot the rolling std (hint)
data["price"].rolling(SMA).std().plot(figsize = (12, 8 ))
plt.show()
```

```python
# Compute lower bound. 
data["Lower"] = data["SMA"] - data["price"].rolling(SMA).std() * dev # Lower Band -2 Std Dev
```

```python
# Compute upper bound
data["Upper"] = data["SMA"] + data["price"].rolling(SMA).std() * dev # Upper Band -2 Std Dev
```

```python
# Plot the bolinger bends
data.drop(columns = "returns").plot(figsize = (12, 8))
plt.show()
```

```python
data.drop(columns = "returns").loc["2019-08"].plot(figsize = (12, 8))
plt.show()
```

# Defining a `Mean-Reversion Strategy` (Bollinger Bands) (Part 2)

```python
data
```

```python
# helper Column
# It's just the different between the Price and its SMA 
data["distance"] = data.price - data.SMA 
```

```python
# 1. oversold -> go long.
# Where the price less than the lower bound, its indicate the the price need go back up to the avarage.
data["position"] = np.where(data.price < data.Lower, 1, np.nan)
```

```python
# 2. overbought -> go short
# Where the price bigger than the upper bound, its inidate that the price need go go back down to the avarage. 
data["position"] = np.where(data.price > data.Upper, -1, data["position"]) 
```

```python
dif1 = -2
dif2 = 3

dif1 * dif2
```

The code uses the logic `data.distance * data.distance.shift(1) < 0` to check whether the sign of the distance (difference between the price and its average) has changed between the current and the previous time step.

If the sign changes, it implies that the price has crossed the SMA, and a position change is considered. In this case, the position is set to 0, indicating a neutral stance.

```python
# 3. crossing SMA ("Middle Band") -> go neutral
# if from long the price cross down the Middle Band (if yes - finish the position)
# if from short the price cross up the Middle Band (if yes - finish the position)

# distance = different between the Price and its avarage.


# So if the price cross SMA (from up to down), 
# its mean that we go nutral (its can be end of short/long position)
# else, we leave the position as is. 

data["position"] = np.where(data.distance * data.distance.shift(1) < 0, 0, data["position"])
```

- Let's say the previous position was a long position (1), and the current conditions do not trigger any of the actions (1-3).
- Without this code, the "position" column might have NaN values, indicating uncertainty about the current position.

```python
# where 1-3 isnÂ´t applicable -> *hold previous position*
data["position"] = data.position.ffill().fillna(0) 
```

# Vectorized Strategy Backtesting

```python
data
```

```python
data["strategy"] = data.position.shift(1) * data["returns"]
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
data["creturns"] = data["returns"].cumsum().apply(np.exp)
data["cstrategy"] = data["strategy"].cumsum().apply(np.exp)
```

```python
data[["creturns", "cstrategy"]].plot(figsize = (12 , 8))
plt.show()
```

```python
data
```

```python
ptc = 0.00007
```

```python
data["trades"] = data.position.diff().fillna(0).abs()
```

```python
data
```

```python
data.trades.value_counts()
```

```python
data["strategy_net"] = data.strategy - data.trades * ptc
```

```python
data["cstrategy_net"] = data.strategy_net.cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns", "cstrategy", "cstrategy_net"]].plot(figsize = (12 , 8))
plt.show()
```

```python
data[["returns", "strategy_net"]].mean() * (4 * 252) # annualized return
```

```python
data[["returns", "strategy_net"]].std() * np.sqrt(4 * 252) # annualized risk
```
