# Combination 

```python
# First, let't take the `positions` based on MACD strtategy
# (and by the way the reutns of the price.. no related to the MACD!)
comb = macd.loc[:, ["returns", "position"]].copy()
```

```python
comb
```

```python
# Rename the column that we'll fmiliar that this is the positions based
# on MACD. 
comb.rename(columns = {"position":"position_MACD"}, inplace = True)
```

```python
# Take the positions based on RSI strategy.
comb["position_RSI"] = rsi.position.astype("int")
```

```python
# So far the combination return the:
# - log return of the instruments. 
# - positions based on MACD strategy standalone
# - positions based on RSI strategy standalone
comb
```

```python
# Combain the positions! 
# We take a position ONLY if both the strategies indicate on the same signal! 
# if both MACD && RSI == 1 ---> BUY! 
# if both MACD && RSI == 0 ---> SELL!
# Else, 0 ---> nutral 
comb["position_comb"] = np.where(comb.position_MACD == comb.position_RSI, comb.position_MACD, 0)
```

```python
comb.position_comb.value_counts()

# See the number of positions of the strategies combination
```

```python
# Plot the positions
# 0  -> nutral
# 1  -> buy
# -1 -> sell
comb.position_comb.plot(figsize = (12, 8))
plt.show()
```

```python
# Compute the reutrn of the combainded strategies
comb["strategy"] = comb["position_comb"].shift(1) * comb["returns"]
```

```python
comb
```

```python
comb.dropna(inplace=True)
```

```python
# Determine the number of trades that we made
# trade = change positions.
comb["trades"] = comb.position_comb.diff().fillna(0).abs()
```

```python
# Reduce the trading costs
comb.strategy = comb.strategy - comb.trades * ptc
```

```python
# cumelative return:

# of buy and hold strategy
comb["creturns"] = comb["returns"].cumsum().apply(np.exp)
# of the combainded strategies (MACD & RSI)
comb["cstrategy"] = comb["strategy"].cumsum().apply(np.exp)
```

```python
# And...plot that
comb[["creturns", "cstrategy"]].plot(figsize = (12, 8), fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

> __Improvment:__ we can make the optimization method on the sombained strategy together, and not seperetly (we run the optimizer for each strategy stand alone).
