# Identifying Local Highs

- We are going to try identigiying local highs.

> It can be powerfull, because if we success to identify the local highs, it can be signal to short position, for exampl.

```python
from scipy.signal import argrelextrema
```

```python
data
```

```python
hh = data.High.copy()
```

```python
hh
```

```python
# 70 -> fing the local high of the last 70 days
order = 70 # approx. 3 month
```

```python
# Run the function, finding the local max's
local_max = argrelextrema(
    hh.values, # the high price dataFrame (in np format) 
    np.greater_equal, # bool operator (>=) 
    order = order # find local high of the last `order` (e.g 70) data points
)  
```

```python
# For the High price data frame, this is the Indexes of 
# local max for each 70 days:
local_max
```

```python
# this is the dates of local max:  
data.index[local_max]
```

```python
# this is the values of local max:
data.High.values[local_max]
```

```python
# Plot the local max and the close price: 
 
data.Close.plot(figsize = (20, 8))
plt.vlines(x = data.index[local_max], ymin = data.Low.min(), ymax = data.High.max(), color = "g")
plt.show()

# we can see that its pretty accurate, and we can developing trading strategy
# based on that!
# e.g: where local maximum: short trigger
```

# Identifying Local Lows

> Similar to identify local high, we can identify local low. It can be trigger to BUY (long)

```python
data
```

```python
# save the low price 
ll = data.Low.copy()
```

```python
# order, is the number of data points that I want to compare the local min
order
```

```python
# Finding the local mins in the 
local_min = argrelextrema(ll.values, np.less_equal, order = order)
```

```python
# the indexes where the local min
local_min
```

```python
# the dates where the local min
data.index[local_min]
```

```python
# the values where the local min
data.Low.values[local_min]
```

__Plot the local min and the local max with order of 70__
- its mean that for each 70 data points we found the local max and the local min.

it can be trading strategy!

but.. can we simply use the code here to create and backtest trading strategies?

__the clear answer is no..__

Why? __look ahead bias.__ 
- Where the algorithem find local min/max, it compare the 70 day in past, and __70 days in future (this is the look-ahead bias)__


But...
> We can just look at the 70 days on __the past__, and determine the local min/max in that dates, and plan the future as well... 

```python

data.Close.plot(figsize = (20, 8))
plt.vlines(x = data.index[local_max], ymin = data.Low.min(), ymax = data.High.max(), color = "g", label = "Local Highs")
plt.vlines(x = data.index[local_min], ymin = data.Low.min(), ymax = data.High.max(), color = "r", label = "Local Lows")
plt.title("Local Highs and Lows for Fibonacci Retracement", fontsize = 15)
plt.legend(fontsize = 13)
plt.show()
```

# High and Lows - an iterative approach

```python
data
```

```python
order
```

__Highs__
- Initial two columns:
    - `hh` for store the local high from range of given dates
    - `hh_date` for store the date on  the hh 

```python
data["hh"] = np.nan
data["hh_date"] = np.nan
```

```python
data
```

```python
# The iterative approch: find the local max of 'n' last day
# its without look ahead bias (its good.)


for bar in range(len(data)): # iterating over the bars
    
    date = data.index[bar] # determine the current barÂ´s date
    
    hh = data.iloc[:bar+1].High # get the High column until current bar
    
    # determine all local highs until current bar
    local_max = argrelextrema(hh.values, np.greater_equal, order = order) 
    
    # determine the most recent local high (price) and add to "hh" column
    data.loc[date, "hh"] = data.High.values[local_max][-1] 
    
    # determine the most recent local high (date) and add to "hh_date" column
    data.loc[date, "hh_date"] = data.index[local_max][-1]  


# Its like to find the "rolling local max"
```

```python
data
```

__Lows__

```python
data["ll"] = np.nan
data["ll_date"] = np.nan
```

```python
# Find the local min, the interative approch

for bar in range(len(data)): # iterating over the bars
    date = data.index[bar] # determine the current barÂ´s date
    ll = data.iloc[:bar+1].Low # get the high column until current bar
    
    # determine all local lows until current bar
    local_min = argrelextrema(ll.values, np.less_equal, order = order)
    
    # determine the most recent local low (price) and add to "ll" column
    data.loc[date, "ll"] = data.Low.values[local_min][-1]
    
    # determine the most recent local low (date) and add to "ll_date" column
    data.loc[date, "ll_date"] = data.index[local_min][-1]


# Its like to find the "rolling local min"
```

# Identifying Trends (Uptrend / Downtrend)

__Lets plot the Close price, rolling local high/low:__

- In green: we have the rolling local max
- In red: we have the rolling local min

```python
data.loc["2010":"2011", ["Close", "hh", "ll"]].plot(figsize = (12, 8), title = "EUR/USD",
                                                    fontsize = 12)
plt.legend(fontsize = 12)
plt.show()


# Interpretation and strategy for trading:
# - if you see that the local min updating more frequency than the local max,
#   it can be a signal to SELL.
#   for ilustration, you can see that at the start, the local min updated
#   more frequency, while the local max stay constrant over long time.
#   its signal to sell! 
# 
# - And of curse, the vica verca:
#   if local max update more frequency while the local min stay constant,
#   it can be signal to BUY
```

__Downtrend__: Most recent lower low is more recent than the most recent higher high <br>
__Uptrend__: Most recent higher high is more recent than the most recent lower low

```python
# defint positions:
# if the *date* of local max higher from the *date* of local min,
# its mean that the local max update frequency, and its signal to BUY (1)
# otherwise, signal to SELL (-1)
data["Trend"] = np.where(data.hh_date > data.ll_date, 1, -1)
```

# A Fibonacci Retracement (23.6%) Breakout Strategy

```python
data
```

__Go Neutral when reaching new Highs/lows (e.g. when Trend reverses)__

```python
data["position"] = np.where(  
    # if the most recent high/low changed, its signal to go neutral
    (data.hh != data.hh.shift()) | (data.ll != data.ll.shift()), 
    0, 
    np.nan
)
```

```python
data.position.value_counts()
```

### Downtrend Decisions (opportunities to take long)

> __Go Long when Price breaks R23.6__

```python
data["position"] = np.where(
    # if the trand is Down
    (data.Trend == "Down") &  
    # and the previos close small then 23.6%
    (data.Close.shift() < data["R23.6"].shift()) & 
    # and then the current close brake cross the 23.6%
    (data.Close > data["R23.6"]), 
    1, # its signal to LONG! the price came from down, 
       # and brake the rescident of the 23.6%!  
    data.position
)
```

> __Go Neutral when Price reaches/breaks R38.2__ (Take Profit)

```python
# if the price cross the 38.2%, its time to take profit! 
# e.g, the price came up from R23.6 to R38.2
data["position"] = np.where(
    # if the trand is down
    (data.Trend == "Down") &
    # and the previos close smaller than the price on the 38.2%  
    (data.Close.shift() < data["R38.2"].shift()) & 
    # and the price now brake up the 38.2% 
    (data.Close >= data["R38.2"]), 
    0, # take profit! its mean that the price probably came from R23.6 to R38.2
    data.position
)
```

> __Go Neutral when Prices reaches/breaks R0__ (Stop Loss)

```python
data["position"] = np.where(
    # If the tread if down
    (data.Trend == "Down") & 
    # and previes close was bigger than the previos low
    (data.Close.shift() > data.ll.shift()) & 
    # and now the Close smaller than low
    (data.Close <= data.ll), 
    0, # its time to stop loss.
    data.position
)
```

```python
data.position.value_counts()
```

### Uptrend Decisions (opportunities to take short)

__Go Short when Price breaks R23.6__

```python
data["position"] = np.where(
    (data.Trend == "Up") & 
    (data.Close.shift() > data["R23.6"].shift()) & 
    (data.Close < data["R23.6"]),
    -1, 
    data.position
)
```

```python
data.position.value_counts()
```

__Go Neutral when Price reaches/breaks R38.2__ (Take profit)

```python
data["position"] = np.where(
    (data.Trend == "Up") & 
    (data.Close.shift() > data["R38.2"].shift()) & 
    (data.Close <= data["R38.2"]), 
    0, 
    data.position
)
```

__Go Neutral when Prices reaches/breaks R0__ (Stop Loss)

```python
data["position"] = np.where(
    (data.Trend == "Up") & 
    (data.Close.shift() < data.hh.shift()) & 
    (data.Close >= data.hh), 
    0, 
    data.position
)
```

__Go Neutral when reaching new Highs/lows (e.g. when Trend reverses)__

```python
data["position"] = np.where(
    (data.hh != data.hh.shift()) | (data.ll != data.ll.shift()), 
    0, 
    data.position
)
```

```python
data["position"] = data.position.ffill()
```

# Vectorized Strategy Backtesting

```python
data
```

```python
data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
data["strategy"] = data.position.shift(1) * data["returns"]
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
ptc = 0.00007
```

```python
data["trades"] = data.position.diff().fillna(0).abs()
```

```python
data.trades.value_counts()
```

```python
data["strategy_net"] = data.strategy - data.trades * ptc
```

```python
data["creturns"] = data["returns"].cumsum().apply(np.exp)
data["cstrategy"] = data["strategy"].cumsum().apply(np.exp)
data["cstrategy_net"] = data["strategy_net"].cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns", "cstrategy", "cstrategy_net"]].plot(figsize = (12 , 8))
plt.show()
```
